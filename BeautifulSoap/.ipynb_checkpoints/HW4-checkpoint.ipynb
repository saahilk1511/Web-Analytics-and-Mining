```python
query = 'COVID-19 and Booster shots'.lower()  
url = f'https://news.google.com/search?q={query}'
```

```python
res = requests.get(url)
soup = bs4.BeautifulSoup(res.text,"lxml")
```

```python
browser.visit(url)
html = browser.html
html
```

```python
soup = bs4.BeautifulSoup(html, 'html.parser')
articles = [] 
for i in range(0,20): 
    a = soup.select('.xrnccd')[i].getText()
    articles.append(a)
    
for i in range(0,len(articles)):
    print("\n",articles[i])
```

```python
df = pd.DataFrame({"Articles": articles})
```

```python
df.head()
```

```python
import nltk
#nltk.download('wordnet')
from nltk import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import LancasterStemmer, WordNetLemmatizer
import gensim
from gensim.test.utils import common_texts
from gensim.models import Word2Vec
from sklearn.manifold import TSNE
import string
import re 
```

```python
from nltk.corpus import stopwords 

stop_words = set(stopwords.words('english') + stopwords.words('spanish'))

lemmer = WordNetLemmatizer()

def preprocess(x):
    # Lower case
    x = x.lower()
    
    # Remove punctuation
    x = re.sub(r'[^\w\s\-]', ' ', x)
   #x = re.sub(r'[^\P{P}\-]+', ' ', x)
    
    
    # Remove numbers
    x = re.sub(r'\d+', '', x)
    
    # Remove stopwords and lemmatize
    x = [lemmer.lemmatize(w) for w in x.split() if w not in stop_words]
    return ' '.join(x) 

df['clean_articles'] = df['Articles'].apply(preprocess)

df.head()
```

```python
import numpy as np
import pandas as pd
from textblob import TextBlob
import sys
import os
import nltk
import re
import string
from nltk.sentiment.vader import SentimentIntensityAnalyzer
#nltk.download('vader_lexicon')
from langdetect import detect
from nltk.stem import SnowballStemmer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from sklearn.feature_extraction.text import CountVectorizer
from afinn import Afinn
```

```python
# Sentiment analysis with AFINN
from afinn import Afinn
afinn = Afinn(emoticons=True)
afinn_scores = [afinn.score(text) for text in df["clean_articles"]]
df['afinn_sentimen'] = afinn_scores

df[['clean_articles','Articles','afinn_sentimen']]
```

```python
 from nltk.sentiment import SentimentIntensityAnalyzer
# Sentiment analysis using nltk_SentimentIntensityAnalyzer

Sent = SentimentIntensityAnalyzer()
SIA = [Sent.polarity_scores(text).get('compound') for text in df["clean_articles"]]
df['nltk_sentiment'] = SIA 


df[['clean_articles','Articles','afinn_sentimen','nltk_sentiment']]
```

```python
from textblob import TextBlob
SIA = [TextBlob(text).sentiment.polarity for text in df["clean_articles"]]
df['TextBlob_sentiment'] = SIA 

df[['clean_articles','Articles','afinn_sentimen','nltk_sentiment','TextBlob_sentiment']]
```

```python
df[['nltk_sentiment','TextBlob_sentiment']] = df[['nltk_sentiment','TextBlob_sentiment']].apply(lambda x: pd.Series.round(x,2)) 
df.head()
```
